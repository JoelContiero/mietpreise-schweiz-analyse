# 02 – Analyse

**Leitfrage:** Wie stark sind die Wohnungsmieten in der Schweiz in den letzten 10–15 Jahren gestiegen — und hielt das mit der Lohn- und Teuerungsentwicklung Schritt?

Datenbasis: `data/processed/miete_loehne_vergleich.csv` (erzeugt in `01_datenaufbereitung.ipynb`).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
v = pd.read_csv("../data/processed/miete_loehne_vergleich.csv")
v.tail()
```

## Auswertung 1: Mieten, Löhne und Teuerung im Vergleich

Alle Reihen auf Basis 2010 = 100. Die Reallöhne berücksichtigen bereits die Teuerung — steigen sie, haben die Löhne die Preise geschlagen.

```python
d = v[v["Jahr"] >= 2005]
plt.figure(figsize=(9, 5))
plt.plot(d["Jahr"], d["Mietindex_idx"], marker="o", label="Mieten")
plt.plot(d["Jahr"], d["Nominallohn_idx"], marker="s", label="Nominallöhne")
plt.plot(d["Jahr"], d["Reallohn_idx"], marker="^", label="Reallöhne")
plt.axhline(100, color="gray", lw=0.8, ls="--")
plt.title("Mieten, Löhne und Teuerung im Vergleich (Schweiz, Basis 2010 = 100)")
plt.xlabel("Jahr"); plt.ylabel("Index (2010 = 100)"); plt.legend()
plt.tight_layout()
plt.savefig("../figures/mieten_vs_loehne.png", dpi=130)
plt.show()
```

_Fazit:_ Über das letzte Jahrzehnt stiegen die Mieten etwas schneller als die Nominallöhne, aber die Reallöhne wuchsen trotzdem — weil die Teuerung in diesem Zeitraum sehr tief war.

## Auswertung 2: Prozentuale Veränderung 2010 → 2020

```python
def veraenderung(spalte, von=2010, bis=2020):
    a = v.loc[v["Jahr"] == von, spalte].values[0]
    b = v.loc[v["Jahr"] == bis, spalte].values[0]
    return round((b / a - 1) * 100, 1)

for spalte, name in [("Mietindex", "Mieten"), ("Nominallohn", "Nominallöhne"), ("Reallohn", "Reallöhne")]:
    print(f"{name:14s}: +{veraenderung(spalte)} %")
```

## Key Findings

- **Mieten 2010–2020: rund +8 %.** Ein spürbarer, aber moderater Anstieg.
- **Mieten stiegen leicht schneller als die Nominallöhne** (+8,2 % vs. +7,2 %).
- **Trotzdem kein realer Verlust:** Die Reallöhne legten am stärksten zu (+8,7 %), weil die Teuerung im Jahrzehnt extrem tief (teils negativ) war. Die Kaufkraft hielt also mit den Mieten Schritt.

> Hinweis: Die Lohndaten enden 2020. Für eine Aktualisierung bis heute müsste eine neuere BFS-Version nachgeladen werden — der Zeitraum, in dem Mieten und Energiepreise ab 2022 stark stiegen, ist hier noch nicht enthalten.
